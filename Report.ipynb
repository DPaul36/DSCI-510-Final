The best takeaway from these results is that baseball is unpredictable. When a team's batting average is .250, they are extremely prone to both hot and cold streaks. Teams that are very consistent, like the Dodgers and Astros, are going to be less prone to these streaks, but teams like the Phillies and the Padres are going to epitomize streaks. From this analysis, we can see that the contact-focused Astros were able to breeze past the teams on these hot streaks by their ability to be persistent. The Dodgers on the other hand, while consistently having the best offense in baseball, did not have this persistent contact-based approach. Pitching analysis however, gives insight into why the Astros gave up fewer runs. The following plots show the results from the season for Spin-rate for the two most commonly thrown pitches, a fastball and a slider.
```python
Pitching_Plots.pitching_plot()
```

Without going into an analysis on pitch usage and the effect that surely has on performance, spin-rate has been a massive focus for teams in the past couple of years. A fastball that has a high spin rate appears to the batter to be rising, while a slider that has a high spin rate "falls off the table". Pair this with the launch-angle and exit-velocity plots, and we can make some very accurate predictions. The results are as follows.

1. The Astros were at the top of the league for spin-rate, while at the bottom of the league for launch-angle. When facing a team like Philadelphia that has a very high team launch angle, the high spin-rate caused the Phillies to pop the ball up for a lot of weak contact.

2. The Phillies had the opposit effect on the Astros, their low spin-rate allowed the Astros to stick with their plan for high-contact rates.

3. The Dodgers pitching wasn't actually all that good. While they posted good results throughout the season, when they faced a team with high launch-angles and exit-velocities in the playoffs, their pitching was described as "flat".
